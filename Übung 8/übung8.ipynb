```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
import random
```

### (a)

```python
def dichtefunktion_fX(x):
    if x > 0 and x < 1:
        return ((12*x**2) * (1 - x))
    else:
        raise ValueError("Only values between greater 0 and  smaller 1 are allowed!")
    
def verteilungsfunktion_FX(x):
    if x > 0 and x < 1:
        return ((4*x**3) - (3*x**4))
    else:
        raise ValueError("Only values between greater 0 and  smaller 1 are allowed!")
```

### (c)

```python
def zufallszahlen(N):
    
    random_numbers = []
    k = 0
    
    min_fX = min([dichtefunktion_fX(x) for x in np.linspace(0.01, 0.999999, 100000)])
    max_fX = max([dichtefunktion_fX(x) for x in np.linspace(0.01, 0.999999, 100000)])

    while k < N:
        X = random.uniform(0, 1)
        Y = random.uniform(min_fX, max_fX)

        if Y < dichtefunktion_fX(X):
            k+=1
            random_numbers.append(dichtefunktion_fX(X))
    return random_numbers
```

### (d)

```python
def empirische_VF(data):
    sorted_data = np.sort(data)
    n = len(sorted_data)
    y = np.arange(1, n + 1) / n
    return sorted_data, y

n = 100
F_Dach = empirische_VF(data=zufallszahlen(n))
F = [verteilungsfunktion_FX(x) for x in np.linspace(0.01, 0.99, n)]
```

```python
plt.title("Künstlich erzeugte Zufallszahlen vs. Theoretisch Verteilung")
plt.step(F_Dach[0], F_Dach[1], where='post', label="F^ (künstlich)")
plt.step(F_Dach[0], F, where='post', label='F (theoretisch)')
plt.legend()
plt.ylabel("Empir.")
plt.grid()
plt.show()
```
