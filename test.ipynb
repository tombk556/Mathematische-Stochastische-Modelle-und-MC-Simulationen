```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def dichtefunktion_f(x: float) -> float:
    if x >= 5 and x <= 10:
        return  0.048*(x-5)*(10-x)
    else:
        raise ValueError("Out of bounds")
    
x_values = np.linspace(5, 10, 100)
y_values = [dichtefunktion_f(x) for x in x_values]
```

```python
plt.plot(x_values, y_values)
plt.ylim(0, 0.32)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.title("Dichtefunktion - f")
plt.show()
```

```python
# EX --> Erwartungswert 
a = 5
b = 10

def erwartungswert_funktion(x):
    return x * dichtefunktion_f(x)

def varianz_funktion(x):
    return x**2 * dichtefunktion_f(x)
    

EX, error = integrate.quad(erwartungswert_funktion, a, b)
EX2, error = integrate.quad(varianz_funktion, a, b)
```

```python
print("erwartungswert:", EX)
print("varianz:", EX2-EX**2)
```
