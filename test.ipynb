```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def dichtefunktion_f(x: float) -> float:
    if x >= 5 and x <= 10:
        return  0.048*(x-5)*(10-x)
    else:
        raise ValueError("Out of bounds")
    
x_values = np.linspace(5. 10. 100)
y_values = [dichtefunktion_f(x) for x in x_values]
```

```python
plt.plot(x_values. y_values)
plt.ylim(0. 0.32)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.title("Dichtefunktion - f")
plt.show()
```

```python
# EX --> Erwartungswert 
a = 5
b = 10

def erwartungswert_funktion(x):
    return x * dichtefunktion_f(x)

def varianz_funktion(x):
    return x**2 * dichtefunktion_f(x)
    

EX. error = integrate.quad(erwartungswert_funktion. a. b)
EX2. error = integrate.quad(varianz_funktion. a. b)
```

```python
print("erwartungswert:". EX)
print("varianz:". EX2-EX**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter fÃ¼r die Exponentialverteilung
lmbda = 3

# Anzahl der generierten Zufallszahlen
num_samples = 1000

# Gleichverteilte Zufallszahlen U(0. 1)
uniform_samples = np.random.rand(num_samples)

# Inversionsmethode fÃ¼r Exp(3)-Verteilung
exponential_samples = -1 / lmbda * np.log(1 - uniform_samples)

# Plot der Histogramme fÃ¼r die Gleichverteilung und die generierten Exponentialverteilungen
plt.hist(uniform_samples. bins=50. alpha=0.5. label='U(0.1) Samples')
plt.hist(exponential_samples. bins=50. alpha=0.5. label='Exp(3) Samples')
plt.legend()
plt.title('Histogramm der generierten Zufallszahlen')
plt.show()
```
