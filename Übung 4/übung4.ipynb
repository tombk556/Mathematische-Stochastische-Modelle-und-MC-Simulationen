```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

### e) Erzeugen von Zufallszahelne die gleichverteilt sind

```python
N = 20
y_values = [random.uniform(0, 1) for _ in range(N)]
```

### f) Berechnenen von x Werten mit Q Funktion

```python
def qfunc(y: float) -> float:
        return (1 - y) ** (-10/13)
x_values = [qfunc(y) for y in y_values]
```

### g) Histogram

```python
plt.hist(x_values, color='blue', alpha=0.7)
plt.xlabel('Werte')
plt.ylabel('HÃ¤ufigkeit')
plt.grid(True)
plt.show()
```

### h) Empirische Verteilungsfunktion

```python
qunatils = []
for i in range(0, len(x_values)):
    lower = 0
    higher = 0
    x = x_values[i]
    for j in x_values:
        if x <= j:
            lower = lower + 1
        if x > j:
            higher = higher + 1
    qunatil = higher / 20
    qunatils.append(qunatil)
```

### i) 

```python
plt.scatter(x_values, qunatils)
plt.xlabel("x")
plt.ylabel("Quantil")
plt.title("Empirische Verteilungsfunktion")
plt.grid(True)
plt.show()
```

```python
def par_func(x):
    return 1 - (1 / x)**1.3
x_values_par = [par_func(x) for x in range(1, 20)]
```

```python
plt.scatter(range(1, 20), x_values_par)
plt.xlabel("x")
plt.ylabel("Pareto Verteilungsfunktion")
plt.title("Pareto Verteilungsfunktion")
plt.grid(True)
plt.show()
```

### j)

```python
values = [qfunc(y) for y in qunatils]
```

```python
plt.scatter(x_values, values)
plt.xlabel("x")
plt.ylabel("Q(F^(x))")
plt.grid(True)
plt.show()
```

```python
from scipy.integrate import quad
import numpy as np

def f(x):
    return x**2 * (1/3) * np.exp(-x * (1/3)) 

result, _ = quad(f, 0, np.inf)

print("Das Ergebnis der Integration von 0 bis unendlich:", result)
```

```python
import sympy as sp

# Symbolische Variable
x = sp.symbols('x')

# Definieren Sie die zu integrierende Funktion
f = (1/3) * sp.exp(-x * (1/3))  # Beispiel: e^x

# Berechnen Sie das unbestimmte Integral (Stammfunktion)
integral = sp.integrate(f, x)

print("Die Stammfunktion von f(x) ist:", integral)
```
