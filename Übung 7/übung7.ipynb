### Generierung der Zufallszahlen

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def zufallszahlen_generator(N: int) -> list:
    k = 1
    random_numbers = []
    while k <= N:
        def verteilung(x):
            y = math.sqrt(x -1)
            return y
        X = random.uniform(1, 5) # X koordinate
        Y = random.uniform(0, 2) # Y koordinate

        if Y < verteilung(X):
            k+=1
            random_numbers.append(verteilung(X))
    return random_numbers

def empirische_cdf(data):
    sorted_data = np.sort(data)
    n = len(sorted_data)
    y = np.arange(1, n + 1) / n
    return sorted_data, y

F_Dach = empirische_cdf(data=zufallszahlen_generator(50))
```

```python
def theoretische_cdf(z):
    if z > 1 and z<5:
        return (1/8)*math.sqrt((z-1)**3)
    else:
        ValueError("Only z values in the range of 1 to 5!!!")

F = [theoretische_cdf(x) for x in np.linspace(1.01, 4.99, 50)]
```

### (a)

```python
plt.title("Empirische CDF vs. Theoretische CDF")
plt.plot(F_Dach[0], label="Empirische CDF")
plt.plot(F, label="Theoretische CDF")
plt.legend()
plt.grid(True)
plt.show()
```
