### Generierung der Zufallszahlen

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def zufallszahlen_generator():
    random_numbers = []
    k = 1

    while k <= 50:
        Uk = random.uniform(1, 5)
        Vk = random.uniform(0, 2)

        if Vk < math.sqrt(Uk - 1):
            random_numbers.append(Uk)
            k += 1
            
    while k <= 50:
        Uk = random.uniform(1, 5)
        Vk = random.uniform(0, 2)

    return random_numbers

F_dach = sorted(zufallszahlen_generator())
```

```python
def F_vonZ(z):
    if z > 1 and z<5:
        return (1/8)*math.sqrt((z-1)**3)
    else:
        ValueError("Only z values in the range of 1 to 5!!!")

F = [F_vonZ(x) for x in np.linspace(1.01, 4.99, 50)]
```

### (a)

```python
plt.title("F^ vs. F")
plt.plot(F_dach, label="F^")
plt.plot(F, label="F")
plt.legend()
plt.grid(True)
plt.show()
```

### (b)

```python
def Quantil(y):
    if y > 0 and y < 1:
        return 1 + 4*y**(2/3)
    else:
        ValueError("Only values in the range of 0 < y < 1")

q_values = [Quantil(y) for y in np.linspace(.01, .99, 50)]
```

```python
plt.title("Quantil vs. Verteilungsfunktion")
plt.plot(q_values, label="Quantil")
plt.plot(F, label="Verteilungsfunktion")
plt.legend()
plt.grid(True)
plt.show()
```
